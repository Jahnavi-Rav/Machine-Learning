```python
y = x*x
```

```python
plt.plot(x, y, 'r*', linestyle = "dashed", linewidth = 2, markersize = 12)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("2D Diagram")
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 4)
plt.plot(x,y, 'go')
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Wave Form")
plt.plot(x, y)
```

```python
sns.pairplot(data);
```
