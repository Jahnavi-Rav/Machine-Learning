```python
import pandas as pd
```

```python
data = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
```

```python
data.head()
```

```python
data.drop("sno", axis = 1, inplace = True)
```

```python
data.head(2)
```

```python
data.shape
```

```python
data.dtypes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(np.arange(len(data['state_name'])),data['positive'], align='center', alpha=0.5,  
                 color=(1,0,0),  
                 edgecolor=(0.5,0.2,0.8))
    
plt.yticks(np.arange(len(data['state_name'])), data['state_name'])  
plt.xlim(1,max(data['positive'])+100) 
plt.xlabel('Positive Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show()  
```

```python
from pandas.io.formats.info import DataFrameInfoPrinter
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(np.arange(len(data['state_name'])),data['new_active'], align='center', alpha=0.5,  
                 color=(1,0.5,0),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(np.arange(len(data['state_name'])), data['state_name'])  
plt.xlabel('Active Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
plt.barh(np.arange(len(data['state_name'])),data['death'], align='center', alpha=0.5,  
                 color=(0,0,1),  
                 edgecolor=(0.5,0.4,0.8)  )
    
plt.yticks(np.arange(len(data['state_name'])), data['state_name'])  
plt.xlim(1,max(data['death'])+10) 
plt.xlabel('Death Number of Cases')  
plt.title('Corona Virus Cases')  
plt.show() 
```
